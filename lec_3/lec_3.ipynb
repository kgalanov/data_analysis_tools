```python
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import PercentFormatter
```

```python
df = pd.read_csv('AirPassengers.csv')
df['Month'] = pd.to_datetime(df['Month'], format='%Y-%m')
df.columns = ['Date', 'Values']
df
```

```python
plt.plot(df.Date, df.Values);
```

```python
a = np.random.normal(loc=5, scale=12)
b = np.random.normal(loc=13, scale=20)
x = np.linspace(0, 4 * np.pi)
y_1 = np.sin(x) * a
y_2 = np.sin(x) * b

plt.plot(x, y_1, label=f'График функции {round(a, 3)}sin(x)')
plt.plot(x, y_2, label=f'График функции {round(b, 3)}sin(x)')
plt.title('Графики периодических функций')
plt.xlabel('Ось Х [0; 4pi]')
plt.ylabel('Ось Y')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df.Date, df.Values, color='yellow', linestyle='--', label='График 1')
plt.scatter(df.Date, df.Values, color='black', marker='o', label='Точки')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
df['Month'] = df['Date'].dt.month
df_bar = df.groupby('Month', as_index=False)['Values'].mean()
plt.bar(df_bar['Month'], df_bar['Values']);
```

```python
plt.figure(figsize=(15, 10))
df['Year'] = df['Date'].dt.year
df_pie = df.groupby('Year', as_index=False)['Values'].sum()
explode = [0] * 11 + [0.1]
plt.pie(df_pie['Values'], labels=df_pie['Year'], autopct='%1.1f', startangle=90, explode=explode);
```

```python
plt.figure(figsize=(12, 8))
N_points = 100000
n_bins = 20
dist1 = np.random.normal(size=N_points)
dist2 = 0.4 * np.random.normal(size=N_points) + 5

fig, axs = plt.subplots(1, 2, tight_layout=True)

N, bins, patches = axs[0].hist(dist1, bins=n_bins)
frac_s = N / N.max()
norm = colors.Normalize(frac_s.min(), frac_s.max())
for this_frac, this_patch in zip(frac_s, patches):
    color = plt.cm.viridis(norm(this_frac))
    this_patch.set_facecolor(color)

axs[1].hist(dist2, bins=n_bins, density=True)
axs[1].yaxis.set_major_formatter(PercentFormatter(xmax=1))
```
