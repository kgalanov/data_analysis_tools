```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('iris')

plt.figure(figsize=(12, 8))
sns.histplot(data, x='petal_length', binwidth=0.25, hue="species", multiple="stack", shrink=0.95)
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency');
```

```python
plt.figure(figsize=(20, 8))
sns.histplot(data, x='petal_length', y='petal_width', bins=15, hue="species", cbar=True)
plt.title('Histogram of Petal')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width');
```

```python
plt.figure(figsize=(12, 8))
sns.kdeplot(data=data, x='sepal_length', bw_adjust=0.6, cut=0, hue="species", fill=True, palette="crest", alpha=0.5,
            linewidth=0)
plt.title('Kernel Density Estimation of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Density');
```

```python
plt.figure(figsize=(12, 8))
sns.kdeplot(data=data, x='sepal_length', y='sepal_width', hue="species", levels=5, thresh=0.2)
plt.title('Kernel Density Estimation of Sepal')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width');
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='petal_length', y='sepal_length', hue="species", palette="deep", style="species",
                size="species", sizes=(50, 150))
plt.title('Scatter Plot with Hue')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length');
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='species', y='petal_length', fill=False, gap=0.2)
plt.title('Boxplot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length');
```

```python
df = sns.load_dataset("titanic")

plt.figure(figsize=(12, 8))
sns.violinplot(data=df, x="class", y="age", hue="alive", split=True, inner="quart", gap=0.05)
plt.title('Violin Plot of Age by Class')
plt.xlabel('Class')
plt.ylabel('Age');
```

```python
plt.figure(figsize=(12, 8))
glue = sns.load_dataset("glue").pivot(index="Model", columns="Task", values="Score")
sns.heatmap(glue, annot=True, fmt="1.1f", linewidth=0.5, cmap=sns.cubehelix_palette(as_cmap=True));
```

```python
data = sns.load_dataset('tips')

sns.catplot(data=data, x='day', y='total_bill', hue='sex', kind='bar')
plt.title('Total Bill by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Total Bill');
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='day', y='total_bill', estimator='sum', data=data)
plt.title('Total Bill by Day (Sum)')
plt.xlabel('Day')
plt.ylabel('Total Bill');
```

```python
data = sns.load_dataset('iris')
correlation_matrix = data[[i for i in data.columns if i != 'species']].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.xlabel('Variables')
plt.ylabel('Variables');
```

```python
sns.jointplot(data=data, x='sepal_width', y='sepal_length', kind='scatter')
plt.title('Jointplot of Sepal Width and Length')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length');
```
